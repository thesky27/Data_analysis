### 6.pandas中的数据处理方法

### 8.自定义函数
apply：对一行货一列做出操作，axis=0的时候队列进行操作
类似于sort方法
每次将dataframe的一行或一列传给function，获取返回值，将返回值放入一个series

```python
titanic_trainval.apply(lambda column:column[99],axis=0)
```

```python
def which_class(row):
    pclass = row['Pclass']
    mes = ['Fisrt','Second','Third']
    if pd.isnull(pclass):
        return 'Unkown'
    else:
        return mes[pclass-1]
titanic_trainval.apply(which_class,axis=1)
```

```python
titanic_trainval = pd.read_csv('./titanic_train.csv')
```

```python
def genetate_age_label(row):
    age = row['Age']
    if pd.isnull(age):
        return 'Unkown'
    elif age<18 :
        return '未成年人'
    else:
        return '成年人'
age_labels = titanic_trainval.apply(genetate_age_label,axis=1)
```

```python
titanic_trainval['Age_label'] = age_labels
age_group_surival = titanic_trainval.pivot_table(index='Age_label',values='Survived')
age_group_surival
```

### Series系列
pandas最核心的结构dataframe

Series就是dataframe的一个分解，dataframe的一行或者是一个

类似于字典

```python
fandango = pd.read_csv('./fandango_score_comparison.csv')
series_file = fandango['FILM'] #列
type(series_file)
```

```python
from pandas import Series
file_names = series_file.values #返回矩阵的所有值
file_names
```

```python
series_rt = fandango['RottenTomatoes']
srt = Series(series_rt.values,index=file_names)
```

```python
rt_critics = Series(fandango['RottenTomatoes'].values,index=fandango['FILM'])
rt_users = Series(fandango['RottenTomatoes_User'].values,index=fandango['FILM'])
rt_mean = (rt_critics+rt_users)/2
rt_mean
```

```python
fandango_films = fandango.set_index('FILM',drop=False)
fandango_films
```

```python
# 切片的操作可以是索引也可以是名称
fandango_films['Avengers: Age of Ultron (2015)':'Top Five (2014)']
fandango_films.iloc[0:5]
```
